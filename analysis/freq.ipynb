```python
from collections import Counter
from transformers import AutoTokenizer
from tqdm.notebook import tqdm
```

```python
tk1 = AutoTokenizer.from_pretrained("/home/gaia_data/iida.h/BEIR/model/pubmed_abst/bert-base-uncased/tokenizer/raw/remove/30522/")
tk2 = AutoTokenizer.from_pretrained("/home/gaia_data/iida.h/BEIR/model/pubmed_abst/bert-base-uncased/tokenizer/raw/remove/33522/")
tk3 = AutoTokenizer.from_pretrained("/home/gaia_data/iida.h/BEIR/model/pubmed_abst/bert-base-uncased/tokenizer/raw/remove/71694//")
```

```python
freq1 = Counter()
freq2 = Counter()
with open("/home/gaia_data/iida.h/pubmed/pubmed_abst_clean.txt") as f:
    for line in tqdm(f):
        t_line1 = tk1.tokenize(line)
        t_line2 = tk2.tokenize(line)
        freq1.update(t_line1)
        freq2.update(t_line2)
        
```

```python
freq3 = Counter()
with open("/home/gaia_data/iida.h/pubmed/pubmed_abst_clean.txt") as f:
    for line in tqdm(f):
        t_line3 = tk3.tokenize(line)
        freq3.update(t_line3)
```

```python
print(len(freq1), len(freq2))
```

```python
def only_topk_rank(freq1, freq2, topk):
    only_tk2 = Counter()
    for k, v in freq2.items():
        if k not in freq1:
            only_tk2[k] = v
    
    only_tk2_top10 = only_tk2.most_common()[:topk]
    only_tk2_top10_rank = []
    tmp = freq1.most_common()
    i = 0
    for w, freq in only_tk2_top10:
        while tmp[i][1] > freq:
            i += 1
        only_tk2_top10_rank.append((w, i))
    return only_tk2_top10_rank
```

```python
only_topk_rank(freq1, freq2, 100)
```

```python
only_topk_rank(freq1, freq3, 1000)
```
